```python
import numpy as np
import pandas as pd
import pyflux as pf
import matplotlib.pyplot as plt
import seaborn
import statsmodels.api as sm
from statsmodels.graphics.api import qqplot
%matplotlib inline 
```

```python
df = pd.read_excel("daily_weather_data_for_Aguafria_2003-2007.xlsx")
df.to_csv("2003.csv", encoding='utf-8', index=False)
```

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d %H:%M:%S')
```

```python
df = pd.read_csv("2003.csv", parse_dates=[0], index_col=0,date_parser=dateparse)
```

```python
df = df.dropna()
```

```python
df
```

```python
dfAmbient = df[["Ambient (°F)"]]
```

```python
dfAmbient
```

```python
seaborn.set()
```

```python
plt.figure(figsize=(20,5))
plt.plot(dfAmbient.index,dfAmbient["Ambient (°F)"])
plt.ylabel("Ambient")
plt.title("Ambient Change");
plt.show()
```

```python
fig = plt.figure(figsize=(20,5))
ax1= fig.add_subplot(111)
diff1 = dfAmbient.diff(1)
diff1.plot(ax=ax1)
```

```python
fig = plt.figure(figsize=(20,5))
ax2= fig.add_subplot(111)
diff2 = dfAmbient.diff(2)
diff2.plot(ax=ax2)
```

```python
fig = plt.figure(figsize=(20,5))
ax3= fig.add_subplot(111)
diff3 = dfAmbient.diff(3)
diff3.plot(ax=ax3)
```

```python
fig = plt.figure(figsize=(20,5))
ax1=fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(diff1,lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(diff1,lags=40,ax=ax2)
```

```python
model = pf.ARIMA(data=dfAmbient,ar=4,ma=4,integ=0,target='Ambient (°F)')
```

```python
x = model.fit("MLE")
x.summary()
```

```python
model.plot_z(indices=range(1,9))
```

```python
model.plot_fit(figsize=(15,5))
```

```python
model.plot_predict_is(50,figsize=(15,5))
```

```python
class Object(object): pass

pd.core.indexes = Object()
pd.core.indexes.datetimes = Object()
pd.core.indexes.datetimes.DatetimeIndex = pd.DatetimeIndex
model.plot_predict(h=50,past_values=20,figsize=(15,5))
```

```python
model.predict(h=50)
```

```python
import numpy as np
import pyflux as pf
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline 

df = pd.read_csv("2003.csv", parse_dates=[0], index_col=0,date_parser=dateparse)
```

```python
dfMultivalue = df[["Ambient (°F)", "Temp(°C)", "E (W/m²)"]]
```

```python
modelMulti = pf.VAR(data=dfMultivalue,lags=2,integ=1)
x = modelMulti.fit()
x.summary()
```

```python
modelMulti.plot_z(list(range(0,6)),figsize=(15,5))
```

```python
modelMulti.plot_fit(figsize=(15,5))
```

```python
class Object(object): pass

pd.core.indexes = Object()
pd.core.indexes.datetimes = Object()
pd.core.indexes.datetimes.DatetimeIndex = pd.DatetimeIndex
modelMulti.plot_predict(past_values=19,h=5,figsize=(15,5))
```

```python
modelMulti.predict(h=5)
```

```python
modelMulti.plot_predict_is(h=30,figsize=((15,5)))
```
